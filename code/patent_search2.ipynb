```python
domain_url_total='http://patft.uspto.gov/netacgi/nph-Parser?Sect1=PTO2&Sect2=HITOFF&u=%2Fnetahtml%2FPTO%2Fsearch-adv.htm&r=0&f=S&l=50&d=PTXT&OS=AN%2F%22google+inc%22+AND+APD%2F20120101-%3E20121231+AND+%22smart+phone%22&RS=%28%28AN%2F%22google+inc%22+AND+APD%2F20120101-%3E20121231%29+AND+%22smart+phone%22%29&Query=AN%2F%22google+inc%22+AND+APD%2F20120101-%3E20121231+AND+%22smart+phone%22&TD=707&Srch1=%28%28%28%22google+inc%22.ASNM.%29+AND+%40AD%3E%3D20120101%3C%3D20121231%29+AND+%22smart+phone%22%29&NextList2=Next+50+Hits'
domain_url2='=Next+50+Hits'
domain_url1=domain_url_total[:len(domain_url_total)-len('=Next+50+Hits')-2]
if (domain_url1[len(domain_url1)-1:]!='t'):
    domain_url1=domain_url_total[:len(domain_url_total)-len('=Next+50+Hits')-1]
print(domain_url1)
#domain_url1='http://patft.uspto.gov/netacgi/nph-Parser?Sect1=PTO2&Sect2=HITOFF&u=%2Fnetahtml%2FPTO%2Fsearch-adv.htm&r=0&f=S&l=50&d=PTXT&OS=AN%2F%22google+inc%22+AND+%22smart+phone%22&RS=AN%2F%22google+inc%22+AND+%22smart+phone%22&Query=AN%2F%22google+inc%22+AND+%22smart+phone%22&TD=2661&Srch1=%22google+inc%22.ASNM.&Srch2=%22smart+phone%22&Conj1=AND&NextList'

import requests
from bs4 import BeautifulSoup
import re
domain='http://patft.uspto.gov'
patent_ary = []
```

```python
def page_process(p):
    # 內文處理
    dic = {}
    dic['patent_no']=href_list[p].text.replace(",",'')
    res1 = requests.get(domain+href_list[p]['href'].replace('&amp;','&'))
    #pos1=res1.text.find('Patent:')
    #dic['patent_no']=res1.text[pos1+7:pos1+15]
         
    pos1=res1.text.find('Current CPC Class:')
    if (pos1<0):
        dic['cpc_class']=''
        pos1=res1.text.find('Current International Class:')
        temp=res1.text[pos1+27:]   
    else:    
        temp=res1.text[pos1+10:]  
        pos1=temp.find('70%')
        temp=temp[pos1+5:]
        pos1=temp.find('</TD')
        #dic['cpc_class']=temp[:pos1].replace('&nbsp','')
        dic['cpc_class']=temp[:pos1].replace('&nbsp','')
        dic['cpc_class']=re.sub("[0-9]{8}","",dic['cpc_class']).replace("()","")
        temp=temp[pos1:]
        pos1=temp.find('Current International Class:')
        
    if (pos1<0):
        dic['international_class']=''
    else:    
        pos1=temp.find('70%')
        temp=temp[pos1+5:]
        pos1=temp.find('</TD')
        #dic['international_class']=temp[:pos1].replace('&nbsp','')
        dic['international_class']=temp[:pos1].replace('&nbsp','')
        dic['international_class']=re.sub("[0-9]{8}","",dic['international_class']).replace("()","")

    patent_ary.append(dic)
```

# the first page

```python
res = requests.get(domain_url1+'1'+domain_url2)
pos=res.text.find("out of") #14
pos1=res.text[pos+15:pos+100].find('strong')  #100隨意給
patent_total=int(res.text[pos+15:(pos+15)+(pos1-2)])
print('total pages={}'.format(patent_total))

pos=res.text.find("<TABLE")
temp=res.text[pos:]
soup = BeautifulSoup(temp, 'html.parser')
href_list=soup.select('a')
for e in range(0,99,2):    
    print(e)
    page_process(e)   
```

# the inner pages

```python
pages_total=int(patent_total/50)+1
for e in range(2,pages_total):   
    print('page {}'.format(e))
    res = requests.get(domain_url1+str(e)+domain_url2)
    soup = BeautifulSoup(res.text, 'html.parser')
    href_list=soup.select('a')
    last_no=110      #10+2*50
    if (last_no>len(href_list)-9):
        last_no=len(href_list)-9 
    for f in range(10,last_no,2):
        print(f)
        page_process(f)
```

# the last page

```python
patent_total=patent_total-50*(pages_total-1)
domain_url3=domain_url1+str(pages_total)+'=Final+'+str(patent_total)+'+Hits'
patent_total=2*patent_total
res = requests.get(domain_url3)

pos=res.text.find("<TABLE")
temp=res.text[pos:]
soup = BeautifulSoup(temp, 'html.parser')
href_list=soup.select('a')
for e in range(0,patent_total,2):
    print(e)
    page_process(e)
```
